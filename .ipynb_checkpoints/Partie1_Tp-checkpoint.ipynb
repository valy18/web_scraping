```python
from selenium.common.exceptions import NoSuchElementException

# Dictionnaire pour stocker les données
data = {
    "Nom_Article": [],
    "Prix_Article": [],
    "Boutique": [],
    "Nombre_Etoiles": [],
    "Lien": []
}



while True:
    # Récupérer la source de la page actuelle
    page = driver.page_source
    soup = BeautifulSoup(page, "html.parser")
    
    # Extraction des données
    article_elements = soup.find_all("article", class_="prd _fb col c-prd")
    for article in article_elements:
        boutik = article.find("div", class_="bdg _mall _xs").text.strip() if article.find("div", class_="bdg _mall _xs") else "Boutique non officielle"
        name = article.find("h3", class_="name").text.strip()
        price = article.find("div", class_="prc").text.strip()
        rating = article.find("div", class_="stars _s").text.split()[0] if article.find("div", class_="stars _s") else "Note non disponible"
        link=url + article.a.next_sibling["href"]

        data["Nom_Article"].append(name)
        data["Prix_Article"].append(price)
        data["Boutique"].append(boutik)
        data["Nombre_Etoiles"].append(rating)
        data["Lien"].append(link)
    
    # Vérification de la présence du bouton "suivant"
    try:
        next_button = driver.find_element(By.CSS_SELECTOR, "a.pg[aria-label='Page suivante']")
        next_page_url = next_button.get_attribute("href")
        driver.get(next_page_url)  # Aller à la page suivante
        time.sleep(4)  # Attendre que la page suivante se charge
    except NoSuchElementException:
        print("Fin de la pagination.")
        break

# Création d'un DataFrame
df1 = pd.DataFrame(data)
df1["Prix_Article"] = df1["Prix_Article"].str.replace("FCFA", "").str.replace(",", "").str.strip().astype(int)

# Affichage des données collectées
print(df1)
```

```python
df1
```

```python
# Filtrer pour garder les articles contenant uniquement des variantes de "Apple" ou "iPhone"
df_filtered = df1[df1["Nom_Article"].str.match(r"(?i)^(.*\bApple\b.*|.*\biPhone\b.*)$", na=False)]
df_filtered

df_filtered["Nom_Article"].unique()
```

```python
# Filtrer les lignes commençant par "Apple" (insensible à la casse)
df_filtered2 = df_filtered[df_filtered["Nom_Article"].str.match(r"(?i)^Apple", na=False)]
df_filtered2
```

```python
df_filtered2= df1[df1["Nom_Article"].str.match(r"(?i)^Apple iPhone", na=False)]
df_filtered2
```

```python
# Uniformiser l'écriture de "Apple iPhone" en "Apple iPhone"
df_filtered2["Nom_Article"] = df_filtered2["Nom_Article"].str.replace(r"(?i)\bApple\s+iphone\b", "Apple iPhone", regex=True)
df_filtered2
```

```python
# Nettoyage pour les rendres conformes 
df_filtered2["Nom_Article"] =df_filtered2["Nom_Article"].str.replace(r"\s?\(.*?\)", "", regex=True)  # Enlever les parenthèses et leur contenu
df_filtered2["Nom_Article"] = df_filtered2["Nom_Article"].str.replace(r"\s?(Black|White|Blue|Red|Gold|Silver|Gray|Space Gray|Noir|Noire|Bleue|Rouge|Jaune|OR|Or|Retina|OLED|Blanc|ROSE|Sable|Sarcelle|NOIRE|Graphite|Rose|Titane|Naturel|Ultramarine|BLANC)$", "", regex=True)  # Enlever les couleurs
df_filtered2["Nom_Article"] = df_filtered2["Nom_Article"].str.replace(r"\s?-+\s?", " ", regex=True)  # Enlever les tirets et les espaces autour
```

```python
# Supprimer les lignes contenant "Super Retina XDR OLED"
df_filtered2 = df_filtered2[~df_filtered2["Nom_Article"].str.contains("Super Retina XDR OLED", case=False, na=False)]
```

```python
# Dataset final sans la colonne 

#Suppression Nombre_Etoiles

df_filtered2 = df_filtered2.drop(columns=["Nombre_Etoiles"])
```
